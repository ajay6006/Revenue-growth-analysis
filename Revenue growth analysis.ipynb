```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_file.csv',encoding='latin1')
```

```python
df.sample(5)
```

```python
# Checking data shape i.e No. of rows and columns
df.shape
```

```python
# Checking descriptive stats
df.describe()
```

```python
df.info()
```

```python
# Converting date columns in a proper format i.e in datetime datatype
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] =pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
# Created new columns for better analysis
df['Order month'] = df['Order Date'].dt.month
df['Order year'] = df['Order Date'].dt.year
df['Order day of week'] = df['Order Date'].dt.dayofweek 
```

```python
df.columns
```

### Q1. Calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales.

```python
sales_by_month =df.groupby('Order month')['Sales'].sum()
sales_by_month
```

```python
plt.figure(figsize=(12,4))
plt.plot(sales_by_month)
plt.title('Sales by Month',size=15)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### Q2. analyze sales based on product categories and determine which category has the lowest sales and which category has the highest sales.

```python
Sales_by_Category = df.groupby('Category')['Sales'].sum()
plt.title('Category wise sales',size=15)
plt.pie(Sales_by_Category, labels=Sales_by_Category.index, autopct='%1.2f%%')
plt.show()
```

### Q.3 sales analysis needs to be done based on sub-categories

```python
Sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum()
Sales_by_subcategory = Sales_by_subcategory.sort_values(ascending=False)  # Sort in descending order
Sales_by_subcategory
```

```python
plt.figure(figsize=(10,4))
plt.title('Sub-Category wise Sales',size=15)
sns.barplot(Sales_by_subcategory)
plt.xticks(rotation=90)
plt.show()
```

### Q.4 analyze the monthly profit from sales and determine which month had the highest profit.

```python
profit_by_month = df.groupby('Order month')['Profit'].sum()
profit_by_month
```

```python
plt.figure(figsize=(12,4))
plt.title('Profit per month',size=15)
sns.lineplot(profit_by_month,color = 'darkblue')
plt.show()
```

### Q.5 Analyze the profit by category and sub-category.

```python
profit_by_Category = df.groupby('Category')['Profit'].sum()
profit_by_Category
```

```python
Profit_by_Category = df.groupby('Category')['Profit'].sum()
plt.title('Category wise Profit',size=15)
plt.pie(Profit_by_Category, labels=Profit_by_Category.index, autopct='%1.2f%%')
plt.show()
```

```python
profit_by_SubCategory = df.groupby('Sub-Category')['Profit'].sum()
profit_by_SubCategory
```

```python
plt.figure(figsize=(10,4))
plt.title('Sub-Category wise Profit',size=15)
sns.barplot(profit_by_SubCategory)
plt.xticks(rotation=90)
plt.show()
```

### Q.6 Analyze the sales to profit ratio by customer segment

```python
Sales_by_segment = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).round(2)
Sales_by_segment
```

```python
Profit_Sales_Ratio = (Sales_by_segment['Sales']/Sales_by_segment['Profit']).round(2)
Profit_Sales_Ratio
```
